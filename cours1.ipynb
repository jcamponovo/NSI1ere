Un nombre ne base 10 s'écrit peut se décomposer en unité, dizaines, centaines etc... Par exmple $257_{10} = 2\times100 + 5\times10 + 7\times1$.  
$100$, $10$ et $1$ sont des puissances de dix car on travaille en base 10 ici. On a donc $257_{10} = 2\times10^2 + 5\times10^1 + 7\times 10^0$ 
  
  
Si on traville en base 2, un nombre pourra se décomposer en unité, "deuzaine", etc... Par exemple $10_2 = 1\times2 + 0\times1 = 1\times2^1 + 0\times2^0$

Autre exemple $11010_2 = 1\times2^4 + 1\times2^3 + 0\times2^2 + 1\times2^1 + 0\times2^0$  

On peut ainsi facilement convertir un nombre de base 2 enbase 10:  
$11010_2 = 1\times16 + 1\times8 + 0\times4 + 1\times2 + 0\times1 = 26_{10}$

**Activité 2:** Convertir un nombre de la base 10 à la base 2  
Pour convertir un nombre en base 2, il faut:
1. tant que le nombre n'est pas nul:
    1. Diviser le nombre par 2 et écrire le reste (0 ou 1) à la gauche de ce qui a déjà été écrit.
    2. Remplacer le nombre par le quotient obtenu lors de la division précédente

  
  
Exemple:  
Le nombre est 13
13 / 2 = 6 reste 1  On écrit 1, le nouveau nombre est 6  
6 / 2 = 3 reste 0  On écrit 0 à la gauche du 1 précédent, on a donc l'écriture 01, le nouveau nombre est 3  
3 / 2 = 1 reste 1  On écrit 1 à gauche, on a donc l'écriture 101, le nouveau nombre est 1  
1 / 2 = 0 reste 1  On écrit 1 à gauche, on a donc l'écriture 1101, le nouveau nombre est 0  
Le nombre est 0, l'algorithme s'arrête.  
On obtient ainsi 1101 en base 2.
