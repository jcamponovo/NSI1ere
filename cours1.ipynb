**Activité 2:** Convertir un nombre de la base 10 à la base 2  
Pour convertir un nombre ne base 2, il faut:
1. Diviser le nombre par 2 et écrire le reste (0 ou 1)
2. Reprendre l'étape 1 avec le quotient obtenu et écrire le reste à la gauche du dernier chiffre.
3. Continuer ainsi jusqu'à obtenir un quotient de 0.
  
  
Exemple:  
13 / 2 = 6 reste 1  
6 / 2 = 3 reste 0  
3 / 2 = 1 reste 1  
1 / 2 = 0 reste 1  
On obtient 1101 en base 2.
